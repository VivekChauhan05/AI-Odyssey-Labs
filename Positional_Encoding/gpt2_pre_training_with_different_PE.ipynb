# Train and Validation data

```python
import os
from tqdm import tqdm
import numpy as np
import tiktoken
from datasets import load_dataset, DatasetDict
import random

# Number of workers for .map() and load_dataset() calls
num_proc = 8
num_proc_load_dataset = num_proc

# Initialize the GPT-2 tokenizer
enc = tiktoken.get_encoding("gpt2")

if __name__ == '__main__':
    # Load the dataset (takes ~54GB in Hugging Face .cache dir, ~8M documents)
    dataset = load_dataset("openwebtext", num_proc=num_proc_load_dataset)["train"]

    # Get the total number of samples
    num_samples = len(dataset)

    # Calculate the sizes for train (10%) and val (0.10%)
    train_size = int(0.10 * num_samples)
    val_size = int(0.0010 * num_samples)

    # Shuffle indices for random sampling
    random.seed(2357)  # For reproducibility
    indices = list(range(num_samples))
    random.shuffle(indices)

    # Select indices for validation and training (non-overlapping)
    val_indices = indices[:val_size]
    train_indices = indices[val_size:val_size + train_size]

    # Create subsets using select
    val = dataset.select(val_indices)
    train = dataset.select(train_indices)

    # Create a DatasetDict with the desired splits
    split_dataset = DatasetDict({'train': train, 'val': val})

    # Define the tokenization function
    def process(example):
        ids = enc.encode_ordinary(example['text'])  # Encode without special tokens
        ids.append(enc.eot_token)  # Add end-of-text token (50256 for GPT-2)
        return {'ids': ids, 'len': len(ids)}

    # Tokenize the dataset
    tokenized = split_dataset.map(
        process,
        remove_columns=['text'],
        desc="tokenizing the splits",
        num_proc=num_proc,
    )

# Define the target directory for saving the files
data_dir = '/content/drive/MyDrive/datasets/input'
os.makedirs(data_dir, exist_ok=True)

# Write tokenized data to binary files
for split, dset in tokenized.items():
    arr_len = np.sum(dset['len'], dtype=np.uint64)
    filename = os.path.join(data_dir, f'{split}.bin')  # Store in the specified directory
    dtype = np.uint16  # Suitable since max token value (50256) < 2^16
    arr = np.memmap(filename, dtype=dtype, mode='w+', shape=(arr_len,))
    total_batches = 1024

    idx = 0
    for batch_idx in tqdm(range(total_batches), desc=f'writing {filename}'):
        batch = dset.shard(num_shards=total_batches, index=batch_idx, contiguous=True).with_format('numpy')
        arr_batch = np.concatenate(batch['ids'])
        arr[idx:idx + len(arr_batch)] = arr_batch
        idx += len(arr_batch)
    arr.flush()

# Resulting files: train.bin and val.bin can be read later with np.memmap
# - val.bin contains your validation data tokens.
```

# Using Absolute Positional Embeddings

```python
import math
import torch
import torch.nn as nn
import tiktoken
import inspect
from dataclasses import dataclass
from torch.nn import functional as F
from google.colab import drive

class LayerNorm(nn.Module):
    """ LayerNorm but with an optional bias. PyTorch doesn't support simply bias=False """

    def __init__(self, ndim, bias):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(ndim))
        self.bias = nn.Parameter(torch.zeros(ndim)) if bias else None

    def forward(self, input):
        return F.layer_norm(input, self.weight.shape, self.weight, self.bias, 1e-5)

# Attention Mechanism
class CausalSelfAttention(nn.Module):

    def __init__(self, config):
        super().__init__()
        assert config.n_embd % config.n_head == 0
        # key, query, value projections for all heads, but in a batch
        self.c_attn = nn.Linear(config.n_embd, 3 * config.n_embd, bias=config.bias) # (768, 2304) ->Linear module stores weight as (2304, 768) and when input is passed the forward pass will run and linear function will be called (output= x*W^T+b) ->  (B, T, 768) @ (768, 2304)
        # output projection
        self.c_proj = nn.Linear(config.n_embd, config.n_embd, bias=config.bias) # (768, 768)
        #self.c_proj.NANOGPT_SCALE_INIT = 1 # Change 1 # This attribute acts as a flag that tells the weight initialization function that special scaling should be applied to this layer.
        # regularization
        self.attn_dropout = nn.Dropout(config.dropout)
        self.resid_dropout = nn.Dropout(config.dropout)
        self.n_head = config.n_head
        self.n_embd = config.n_embd
        self.dropout = config.dropout
        # flash attention make GPU go brrrrr but support is only in PyTorch >= 2.0
        self.flash = hasattr(torch.nn.functional, 'scaled_dot_product_attention')
        if not self.flash:
            print("WARNING: using slow attention. Flash Attention requires PyTorch >= 2.0")
            # causal mask to ensure that attention is only applied to the left in the input sequence
            self.register_buffer("bias", torch.tril(torch.ones(config.block_size, config.block_size))
                                        .view(1, 1, config.block_size, config.block_size))

    def forward(self, x):
        B, T, C = x.size() # batch size, sequence length, embedding dimensionality (n_embd) (B, T, 768)
        # calculate query, key, values for all heads in batch and move head forward to be the batch dim
        # nh is "number of heads", hs is "head size", and C (number of channels) = nh * hs
        # e.g. in GPT-2 (124M), n_head=12, hs=64, so nh*hs=C=768 channels in the Transformer
        qkv = self.c_attn(x)  # (B, T, 2304)
        q, k, v = qkv.split(self.n_embd, dim=2) # q-> (B, T, 768), k-> (B, T, 768), v-> (B, T, 768)
        k = k.view(B, T, self.n_head, C // self.n_head).transpose(1, 2) # (B, nh, T, hs)-> (B, 12, T, 64)
        q = q.view(B, T, self.n_head, C // self.n_head).transpose(1, 2) # (B, nh, T, hs)-> (B, 12, T, 64)
        v = v.view(B, T, self.n_head, C // self.n_head).transpose(1, 2) # (B, nh, T, hs)-> (B, 12, T, 64)

        # causal self-attention; Self-attend: (B, nh, T, hs) x (B, nh, hs, T) -> (B, nh, T, T)
        if self.flash:
            # efficient attention using Flash Attention CUDA kernels
            y = torch.nn.functional.scaled_dot_product_attention(q, k, v, attn_mask=None, dropout_p=self.dropout if self.training else 0, is_causal=True)
        else:
            # manual implementation of attention
            att = (q @ k.transpose(-2, -1)) * (1.0 / math.sqrt(k.size(-1)))
            att = att.masked_fill(self.bias[:,:,:T,:T] == 0, float('-inf'))
            att = F.softmax(att, dim=-1)
            att = self.attn_dropout(att)
            y = att @ v # (B, nh, T, T) x (B, nh, T, hs) -> (B, nh, T, hs)
        y = y.transpose(1, 2).contiguous().view(B, T, C) # re-assemble all head outputs side by side

        # output projection
        y = self.resid_dropout(self.c_proj(y))
        return y


# FeedForward Network
class MLP(nn.Module):  # Define the MLP class, a simple Multi-Layer Perceptron, as a subclass of PyTorch's nn.Module.

    def __init__(self, config):  # Constructor method that takes a config object containing model hyperparameters.
        super().__init__()  # Initialize the parent nn.Module class.
        self.c_fc   = nn.Linear(config.n_embd, 4 * config.n_embd, bias=config.bias)  # Define the first linear layer that expands the embedding size by 4x.
        self.gelu   = nn.GELU(approximate='tanh')  # Define the GELU activation function using the 'tanh' approximation for non-linearity.
        self.c_proj = nn.Linear(4 * config.n_embd, config.n_embd, bias=config.bias)  # Define the second linear layer that projects back to the original embedding size.
        #self.c_proj.NANOGPT_SCALE_INIT = 1  # This attribute acts as a flag that tells the weight initialization function that special scaling should be applied to this layer.
        self.dropout = nn.Dropout(config.dropout)

    def forward(self, x):  # Define the forward pass method for processing input x.
        x = self.c_fc(x)  # Pass the input through the first linear layer.
        x = self.gelu(x)  # Apply the GELU activation function to introduce non-linearity.
        x = self.c_proj(x)  # Pass the activated output through the second linear layer (fixed to call the layer properly).
        x = self.dropout(x)
        return x  # Return the final output.


# Transformer Block
class Block(nn.Module):

    def __init__(self, config):
        super().__init__()  # Initialize the base nn.Module.
        self.ln_1 = LayerNorm(config.n_embd, bias=config.bias)  # Initialize the first layer normalization layer for the input, normalizing over the embedding dimension (config.n_embd).
        self.attn = CausalSelfAttention(config)  # Create the causal self-attention module, which will compute attention with masking to prevent attending to future tokens.
        self.ln_2 = LayerNorm(config.n_embd, bias=config.bias)  # Initialize the second layer normalization layer,to normalize the output before passing it to the MLP.
        self.mlp = MLP(config)  # Create the feed-forward network (MLP) module that processes the normalized output from the attention block.

    def forward(self, x):
        # 'x' is the input tensor representing token embeddings.
        x = x + self.attn(self.ln_1(x))
        # 1. Apply layer normalization (ln_1) to 'x'.
        # 2. Pass the normalized tensor through the causal self-attention module.
        # 3. Add the attention output back to the original 'x' (residual connection).
        x = x + self.mlp(self.ln_2(x))
        # 1. Apply the second layer normalization (ln_2) to the current 'x'.
        # 2. Pass the normalized tensor through the MLP (feed-forward network).
        # 3. Add the MLP output to 'x' (another residual connection).
        return x

# Define a configuration class for the GPT model using dataclass for clarity and simplicity
@dataclass
class GPTConfig:
    block_size: int = 1024    # Maximum number of tokens per input sequence
    vocab_size: int = 50304   # number of tokens: 50,000 BPE merges + 256 bytes tokens + 1 <|endoftext|> token
    n_layer: int = 12         # Number of transformer blocks (layers)
    n_head: int = 12          # Number of attention heads per transformer block
    n_embd: int = 768         # Dimensionality of the token embeddings
    dropout: float = 0.0
    bias: bool = True # True: bias in Linears and LayerNorms, like GPT-2. False: a bit better and faster

# Define the GPT model class, inheriting from PyTorch's nn.Module
class GPT(nn.Module):

    def __init__(self, config):
        super().__init__() # Initialize the parent nn.Module class
        assert config.vocab_size is not None
        assert config.block_size is not None
        self.config = config # Save the configuration for later use

        # Create a dictionary of sub-modules that make up the transformer
        self.transformer = nn.ModuleDict(dict(
            wte = nn.Embedding(config.vocab_size, config.n_embd), # Token Embeddings: Maps token IDs to dense vector representations  (50257, 768)
            wpe = nn.Embedding(config.block_size, config.n_embd), # Positional Encodings: Provides positional information to the model (1024, 768)
            drop = nn.Dropout(config.dropout),
            h = nn.ModuleList([Block(config) for _ in range(config.n_layer)]), # Transformer Blocks: A list of 'Block' modules (each is a transformer block)  12 blocks in GPT-2
            ln_f = LayerNorm(config.n_embd, bias=config.bias), # Final Layer Normalization: Normalizes the output of the transformer stack
        ))
        self.lm_head = nn.Linear(config.n_embd, config.vocab_size, bias=False) # Language Modeling Head: A linear layer that projects embeddings back to vocabulary size for predictions

        self.transformer.wte.weight  = self.lm_head.weight

        # init params
        self.apply(self._init_weights)

        # apply special scaled init to the residual projections, per GPT-2 paper
        for pn, p in self.named_parameters():
            if pn.endswith('c_proj.weight'):
                torch.nn.init.normal_(p, mean=0.0, std=0.02/math.sqrt(2 * config.n_layer))

        # report number of parameters
        print("number of parameters: %.2fM" % (self.get_num_params()/1e6,))


    def get_num_params(self, non_embedding=True):
        """
        Return the number of parameters in the model.
        For non-embedding count (default), the position embeddings get subtracted.
        The token embeddings would too, except due to the parameter sharing these
        params are actually used as weights in the final layer, so we include them.
        """
        n_params = sum(p.numel() for p in self.parameters())
        if non_embedding:
            n_params -= self.transformer.wpe.weight.numel()
        return n_params


    def _init_weights(self, module):
        # Check if the module is a Linear layer (fully connected layer)
        if isinstance(module, nn.Linear):
            # Initialize the weight matrix with values from a normal distribution
            # Mean = 0.0 and standard deviation = 0.02
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)

            # If the Linear layer has a bias term, initialize it to zeros
            if module.bias is not None:
                torch.nn.init.zeros_(module.bias)

        # Check if the module is an Embedding layer
        elif isinstance(module, nn.Embedding):
            # Initialize the embedding weights with values from a normal distribution
            # Mean = 0.0 and standard deviation = 0.02
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)


    def forward(self, idx, targets=None):
        device = idx.device
        # idx is of shape (B, T)
        B, T = idx.size()
        assert T <= self.config.block_size, f"Cannot forward sequence of lenght {T}, block size is only {self.config.block_size}"
        pos = torch.arange(0, T, dtype=torch.long, device=device) # shape (T)

        # forward the token and position embeddings
        pos_emb = self.transformer.wpe(pos) # position embeddings of shape (T, n_embd)-> (T, 768)
        tok_emb = self.transformer.wte(idx) # token embeddings of shape (B, T, n_embd)-> (B, T, 768)
        x = self.transformer.drop(tok_emb + pos_emb) # (B, T, 768) + (B, T, 768)
        # forward the blocks of the transformer
        for block in self.transformer.h:
          x = block(x)
        # forward the final layernorm and the classifier
        x = self.transformer.ln_f(x)
        loss = None
        if targets is not None:
            # Reshape logits from (B, T, vocab_size) to (B*T, vocab_size) and targets from (B, T) to (B*T,)
            # so that each token's prediction is treated as an independent example.
            # Then, F.cross_entropy computes the average cross-entropy loss over all these tokens,
            # comparing the predicted probability distribution (from logits) to the true token IDs (targets).
            logits = self.lm_head(x)
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))
        else:
            # inference-time mini-optimization: only forward the lm_head on the very last position
            logits = self.lm_head(x[:, [-1], :]) # note: using list [-1] to preserve the time dim
            loss = None

        return logits, loss


    @classmethod
    def from_pretrained(cls, model_type):
        """Loads pretrained GPT-2 model weights from huggingface"""
        assert model_type in {'gpt2', 'gpt-medium', 'gpt2-large', 'gpt2-xl'}
        from transformers import GPT2LMHeadModel
        print("loading weights from pretrained gpt: %s" % model_type )

        # n_layer, n_head and n_embd are determined form model_type
        config_args = {
            'gpt2':       dict(n_layer=12, n_head=12, n_embd=768), # 124M params
            'gpt-medium': dict(n_layer=24, n_head=16, n_embd=1024), # 350M params
            'gpt2-large':dict(n_layer=36, n_head=20, n_embd=1280), # 774M params
            'gpt2-xl': dict(n_layer=48, n_head=25, n_embd=1600), # 1558M params
        }[model_type]
        print("forcing vocab_size=50257, block_size=512, bias=True")
        config_args['vocab_size'] = 50257 # always 50257 for GPT model checkpoints
        config_args['block_size'] = 512 # always 1024 for GPT model checkpoints
        config_args['bias'] = True # always True for GPT model checkpoints
        # create a from-scratch initialized minGPT model
        config = GPTConfig(**config_args)
        model = GPT(config)
        sd = model.state_dict()
        sd_keys = sd.keys()
        sd_keys = [k for k in sd_keys if not k.endswith('.attn.bias')] # discard this mask/ buffer, not a param


        # init as huggingface.transformers model
        model_hf = GPT2LMHeadModel.from_pretrained(model_type)
        sd_hf = model_hf.state_dict()

        # copy while ensuring all of the parameters are aligned and match in names and shapes
        sd_keys_hf = sd_hf.keys()
        sd_keys_hf = [k for k in sd_keys_hf if not k.endswith('.attn.masked_bias')] # ignore these, just a buffer
        sd_keys_hf = [k for k in sd_keys_hf if not k.endswith('.attn.bias')] # same, just the mask (buffer)
        transposed = ['attn.c_attn.weight', 'attn.c_proj.weight', 'mlp.c_fc.weight', 'mlp.c_proj.weight']
        # basically the openai checkpoints use a "Conv1D" module, but we only want to use a vanilla Linear
        # this means that we have to transpose these weights when we import them
        assert len(sd_keys_hf) == len(sd_keys), f"mismatched keys: {len(sd_keys_hf)} != {len(sd_keys)}"
        for k in sd_keys_hf:
            if any(k.endswith(w) for w in transposed):
                # special treatment for the Conv1D weights we need to transpose
                assert sd_hf[k].shape[::-1] == sd[k].shape
                with torch.no_grad():
                   sd[k].copy_(sd_hf[k].t())
            else:
               # vanilla copy over the other parameters
               assert sd_hf[k].shape == sd[k].shape
               with torch.no_grad():
                   sd[k].copy_(sd_hf[k])
        return model

    def configure_optimizers(self, weight_decay, learning_rate,  betas, device_type):
        # start with all of the candidate parameters (that require grad)
        param_dict = {pn: p for pn, p in self.named_parameters()}
        param_dict = {pn:p for pn, p in param_dict.items() if p.requires_grad}
        # create optim groups. Any parameters that is 2D will be weight delayed, otherwise no.
        # i.e. all weight tensors in matmuls + embeddings decay, all biases and layernorms don't.
        decay_params = [p for n, p in param_dict.items() if p.dim() >= 2]
        nodecay_params = [p for n, p in param_dict.items() if p.dim() < 2]
        optim_groups = [
            {'params': decay_params, 'weight_decay': weight_decay},
            {'params': nodecay_params, 'weight_decay': 0.0}
        ]
        num_decay_params = sum(p.numel() for p in decay_params)
        num_nodecay_params = sum(p.numel() for p in nodecay_params)
        print(f"num decayed parameter tensors: {len(decay_params)}, with {num_decay_params:,} parameters")
        print(f"num non-decayed parameter tensors: {len(nodecay_params)}, with {num_nodecay_params:,} parameters")
        # Create AdamW optimizer and use the fused version if it is available
        fused_available = 'fused' in inspect.signature(torch.optim.AdamW).parameters
        use_fused = fused_available and device_type == 'cuda'
        print(f"using fused AdamW: {use_fused}")
        optimizer = torch.optim.AdamW(optim_groups, lr=learning_rate, betas=(0.9, 0.95), eps = 1e-8, fused=use_fused)
        print(f"using fused AdamW: {use_fused}")
        return optimizer
```

# Using NoPE

```python
import math
import torch
import torch.nn as nn
import tiktoken
import inspect
from dataclasses import dataclass
from torch.nn import functional as F
from google.colab import drive

class LayerNorm_NoPE(nn.Module):
    """ LayerNorm but with an optional bias. PyTorch doesn't support simply bias=False """
    def __init__(self, ndim, bias):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(ndim))
        self.bias = nn.Parameter(torch.zeros(ndim)) if bias else None

    def forward(self, input):
        return F.layer_norm(input, self.weight.shape, self.weight, self.bias, 1e-5)

class CausalSelfAttention_NoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        assert config.n_embd % config.n_head == 0
        self.c_attn = nn.Linear(config.n_embd, 3 * config.n_embd, bias=config.bias)
        self.c_proj = nn.Linear(config.n_embd, config.n_embd, bias=config.bias)
        self.attn_dropout = nn.Dropout(config.dropout)
        self.resid_dropout = nn.Dropout(config.dropout)
        self.n_head = config.n_head
        self.n_embd = config.n_embd
        self.dropout = config.dropout
        self.flash = hasattr(torch.nn.functional, 'scaled_dot_product_attention')
        if not self.flash:
            print("WARNING: using slow attention. Flash Attention requires PyTorch >= 2.0")
            self.register_buffer("bias", torch.tril(torch.ones(config.block_size, config.block_size))
                                        .view(1, 1, config.block_size, config.block_size))

    def forward(self, x):
        B, T, C = x.size()
        qkv = self.c_attn(x)
        q, k, v = qkv.split(self.n_embd, dim=2)
        k = k.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)
        q = q.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)
        v = v.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)
        if self.flash:
            y = torch.nn.functional.scaled_dot_product_attention(q, k, v, attn_mask=None, dropout_p=self.dropout if self.training else 0, is_causal=True)
        else:
            att = (q @ k.transpose(-2, -1)) * (1.0 / math.sqrt(k.size(-1)))
            att = att.masked_fill(self.bias[:,:,:T,:T] == 0, float('-inf'))
            att = F.softmax(att, dim=-1)
            att = self.attn_dropout(att)
            y = att @ v
        y = y.transpose(1, 2).contiguous().view(B, T, C)
        y = self.resid_dropout(self.c_proj(y))
        return y

class MLP_NoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.c_fc   = nn.Linear(config.n_embd, 4 * config.n_embd, bias=config.bias)
        self.gelu   = nn.GELU(approximate='tanh')
        self.c_proj = nn.Linear(4 * config.n_embd, config.n_embd, bias=config.bias)
        self.dropout = nn.Dropout(config.dropout)

    def forward(self, x):
        x = self.c_fc(x)
        x = self.gelu(x)
        x = self.c_proj(x)
        x = self.dropout(x)
        return x

class Block_NoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.ln_1 = LayerNorm_NoPE(config.n_embd, bias=config.bias)
        self.attn = CausalSelfAttention_NoPE(config)
        self.ln_2 = LayerNorm_NoPE(config.n_embd, bias=config.bias)
        self.mlp = MLP_NoPE(config)

    def forward(self, x):
        x = x + self.attn(self.ln_1(x))
        x = x + self.mlp(self.ln_2(x))
        return x

@dataclass
class GPTConfig_NoPE:
    block_size: int = 1024
    vocab_size: int = 50304
    n_layer: int = 12
    n_head: int = 12
    n_embd: int = 768
    dropout: float = 0.0
    bias: bool = True

class GPT_NoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        assert config.vocab_size is not None
        assert config.block_size is not None
        self.config = config

        self.transformer = nn.ModuleDict(dict(
            wte = nn.Embedding(config.vocab_size, config.n_embd),
            drop = nn.Dropout(config.dropout),
            h = nn.ModuleList([Block_NoPE(config) for _ in range(config.n_layer)]),
            ln_f = LayerNorm_NoPE(config.n_embd, bias=config.bias),
        ))
        self.lm_head = nn.Linear(config.n_embd, config.vocab_size, bias=False)
        self.transformer.wte.weight = self.lm_head.weight

        self.apply(self._init_weights)
        for pn, p in self.named_parameters():
            if pn.endswith('c_proj.weight'):
                torch.nn.init.normal_(p, mean=0.0, std=0.02/math.sqrt(2 * config.n_layer))

        print("number of parameters: %.2fM" % (self.get_num_params()/1e6,))

    def get_num_params(self, non_embedding=True):
        n_params = sum(p.numel() for p in self.parameters())
        # No wpe to subtract, so return total parameters directly
        return n_params

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                torch.nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, idx, targets=None):
        device = idx.device
        B, T = idx.size()
        assert T <= self.config.block_size, f"Cannot forward sequence of length {T}, block size is only {self.config.block_size}"

        # Only token embeddings, no positional embeddings
        tok_emb = self.transformer.wte(idx)  # (B, T, n_embd)
        x = self.transformer.drop(tok_emb)

        for block in self.transformer.h:
            x = block(x)
        x = self.transformer.ln_f(x)
        if targets is not None:
            logits = self.lm_head(x)
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))
        else:
            logits = self.lm_head(x[:, [-1], :])
            loss = None
        return logits, loss

    @classmethod
    def from_pretrained(cls, model_type):
        assert model_type in {'gpt2', 'gpt-medium', 'gpt2-large', 'gpt2-xl'}
        from transformers import GPT2LMHeadModel
        print("loading weights from pretrained gpt: %s" % model_type)

        config_args = {
            'gpt2':       dict(n_layer=12, n_head=12, n_embd=768),
            'gpt-medium': dict(n_layer=24, n_head=16, n_embd=1024),
            'gpt2-large': dict(n_layer=36, n_head=20, n_embd=1280),
            'gpt2-xl':    dict(n_layer=48, n_head=25, n_embd=1600),
        }[model_type]
        config_args['vocab_size'] = 50257
        config_args['block_size'] = 512
        config_args['bias'] = True
        config = GPTConfig(**config_args)
        model = GPT(config)
        sd = model.state_dict()
        sd_keys = [k for k in sd.keys() if not k.endswith('.attn.bias')]

        model_hf = GPT2LMHeadModel.from_pretrained(model_type)
        sd_hf = model_hf.state_dict()
        sd_keys_hf = [k for k in sd_hf.keys() if not k.endswith('.attn.masked_bias') and not k.endswith('.attn.bias')]
        transposed = ['attn.c_attn.weight', 'attn.c_proj.weight', 'mlp.c_fc.weight', 'mlp.c_proj.weight']
        assert len(sd_keys_hf) == len(sd_keys), f"mismatched keys: {len(sd_keys_hf)} != {len(sd_keys)}"
        for k in sd_keys_hf:
            if any(k.endswith(w) for w in transposed):
                assert sd_hf[k].shape[::-1] == sd[k].shape
                with torch.no_grad():
                    sd[k].copy_(sd_hf[k].t())
            else:
                assert sd_hf[k].shape == sd[k].shape
                with torch.no_grad():
                    sd[k].copy_(sd_hf[k])
        return model

    def configure_optimizers(self, weight_decay, learning_rate, betas, device_type):
        param_dict = {pn: p for pn, p in self.named_parameters() if p.requires_grad}
        decay_params = [p for n, p in param_dict.items() if p.dim() >= 2]
        nodecay_params = [p for n, p in param_dict.items() if p.dim() < 2]
        optim_groups = [
            {'params': decay_params, 'weight_decay': weight_decay},
            {'params': nodecay_params, 'weight_decay': 0.0}
        ]
        num_decay_params = sum(p.numel() for p in decay_params)
        num_nodecay_params = sum(p.numel() for p in nodecay_params)
        print(f"num decayed parameter tensors: {len(decay_params)}, with {num_decay_params:,} parameters")
        print(f"num non-decayed parameter tensors: {len(nodecay_params)}, with {num_nodecay_params:,} parameters")
        fused_available = 'fused' in inspect.signature(torch.optim.AdamW).parameters
        use_fused = fused_available and device_type == 'cuda'
        print(f"using fused AdamW: {use_fused}")
        optimizer = torch.optim.AdamW(optim_groups, lr=learning_rate, betas=betas, eps=1e-8, fused=use_fused)
        return optimizer
```

```python
import math
import torch
import torch.nn as nn
import tiktoken
import inspect
from dataclasses import dataclass
from torch.nn import functional as F
from google.colab import drive

class LayerNorm_RoPE(nn.Module):
    """ LayerNorm but with an optional bias. PyTorch doesn't support simply bias=False """
    def __init__(self, ndim, bias):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(ndim))
        self.bias = nn.Parameter(torch.zeros(ndim)) if bias else None

    def forward(self, input):
        return F.layer_norm(input, self.weight.shape, self.weight, self.bias, 1e-5)

class CausalSelfAttention_RoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        assert config.n_embd % config.n_head == 0
        self.c_attn = nn.Linear(config.n_embd, 3 * config.n_embd, bias=config.bias)
        self.c_proj = nn.Linear(config.n_embd, config.n_embd, bias=config.bias)
        self.attn_dropout = nn.Dropout(config.dropout)
        self.resid_dropout = nn.Dropout(config.dropout)
        self.n_head = config.n_head
        self.n_embd = config.n_embd
        self.dropout = config.dropout
        self.flash = hasattr(torch.nn.functional, 'scaled_dot_product_attention')
        if not self.flash:
            print("WARNING: using slow attention. Flash Attention requires PyTorch >= 2.0")
            self.register_buffer("bias", torch.tril(torch.ones(config.block_size, config.block_size))
                                        .view(1, 1, config.block_size, config.block_size))

    def forward(self, x):
        B, T, C = x.size()
        qkv = self.c_attn(x)
        q, k, v = qkv.split(self.n_embd, dim=2)
        k = k.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)
        q = q.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)
        v = v.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)

        # Apply rotary embeddings
        q, k = self.apply_rotary_emb(q, k, T)

        if self.flash:
            y = torch.nn.functional.scaled_dot_product_attention(q, k, v, attn_mask=None, dropout_p=self.dropout if self.training else 0, is_causal=True)
        else:
            att = (q @ k.transpose(-2, -1)) * (1.0 / math.sqrt(k.size(-1)))
            att = att.masked_fill(self.bias[:,:,:T,:T] == 0, float('-inf'))
            att = F.softmax(att, dim=-1)
            att = self.attn_dropout(att)
            y = att @ v
        y = y.transpose(1, 2).contiguous().view(B, T, C)
        y = self.resid_dropout(self.c_proj(y))
        return y

    def apply_rotary_emb(self, q, k, T):
        """Apply rotary position embeddings to query and key vectors."""
        device = q.device
        hs = self.n_embd // self.n_head  # Head size
        # Compute frequencies: theta_i = 10000^(-2i/hs) for i in [0, hs//2 - 1]
        theta = torch.pow(10000, -torch.arange(0, hs, 2, device=device) / hs)  # (hs//2,)
        m = torch.arange(T, device=device).unsqueeze(1)  # (T, 1)
        angles = m * theta  # (T, hs//2)
        cos_vals = torch.cos(angles)  # (T, hs//2)
        sin_vals = torch.sin(angles)  # (T, hs//2)

        # Split q and k into even and odd indices
        q_even = q[:, :, :, 0::2]  # (B, nh, T, hs//2)
        q_odd = q[:, :, :, 1::2]   # (B, nh, T, hs//2)
        k_even = k[:, :, :, 0::2]  # (B, nh, T, hs//2)
        k_odd = k[:, :, :, 1::2]   # (B, nh, T, hs//2)

        # Apply rotation: q'_even = q_even * cos - q_odd * sin, q'_odd = q_even * sin + q_odd * cos
        q_rot_even = q_even * cos_vals - q_odd * sin_vals
        q_rot_odd = q_even * sin_vals + q_odd * cos_vals
        k_rot_even = k_even * cos_vals - k_odd * sin_vals
        k_rot_odd = k_even * sin_vals + k_odd * cos_vals

        # Reconstruct q and k by interleaving even and odd parts
        q_rot = torch.zeros_like(q)
        k_rot = torch.zeros_like(k)
        q_rot[:, :, :, 0::2] = q_rot_even
        q_rot[:, :, :, 1::2] = q_rot_odd
        k_rot[:, :, :, 0::2] = k_rot_even
        k_rot[:, :, :, 1::2] = k_rot_odd

        return q_rot, k_rot

class MLP_RoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.c_fc   = nn.Linear(config.n_embd, 4 * config.n_embd, bias=config.bias)
        self.gelu   = nn.GELU(approximate='tanh')
        self.c_proj = nn.Linear(4 * config.n_embd, config.n_embd, bias=config.bias)
        self.dropout = nn.Dropout(config.dropout)

    def forward(self, x):
        x = self.c_fc(x)
        x = self.gelu(x)
        x = self.c_proj(x)
        x = self.dropout(x)
        return x

class Block_RoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.ln_1 = LayerNorm_RoPE(config.n_embd, bias=config.bias)
        self.attn = CausalSelfAttention_RoPE(config)
        self.ln_2 = LayerNorm_RoPE(config.n_embd, bias=config.bias)
        self.mlp = MLP_RoPE(config)

    def forward(self, x):
        x = x + self.attn(self.ln_1(x))
        x = x + self.mlp(self.ln_2(x))
        return x

@dataclass
class GPTConfig_RoPE:
    block_size: int = 1024
    vocab_size: int = 50304
    n_layer: int = 12
    n_head: int = 12
    n_embd: int = 768
    dropout: float = 0.0
    bias: bool = True

class GPT_RoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        assert config.vocab_size is not None
        assert config.block_size is not None
        self.config = config

        self.transformer = nn.ModuleDict(dict(
            wte = nn.Embedding(config.vocab_size, config.n_embd),
            drop = nn.Dropout(config.dropout),
            h = nn.ModuleList([Block_RoPE(config) for _ in range(config.n_layer)]),
            ln_f = LayerNorm_RoPE(config.n_embd, bias=config.bias),
        ))
        self.lm_head = nn.Linear(config.n_embd, config.vocab_size, bias=False)
        self.transformer.wte.weight = self.lm_head.weight

        self.apply(self._init_weights)
        for pn, p in self.named_parameters():
            if pn.endswith('c_proj.weight'):
                torch.nn.init.normal_(p, mean=0.0, std=0.02/math.sqrt(2 * config.n_layer))

        print("number of parameters: %.2fM" % (self.get_num_params()/1e6,))

    def get_num_params(self, non_embedding=True):
        n_params = sum(p.numel() for p in self.parameters())
        return n_params

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                torch.nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, idx, targets=None):
        device = idx.device
        B, T = idx.size()
        assert T <= self.config.block_size, f"Cannot forward sequence of length {T}, block size is only {self.config.block_size}"

        tok_emb = self.transformer.wte(idx)  # (B, T, n_embd)
        x = self.transformer.drop(tok_emb)

        for block in self.transformer.h:
            x = block(x)
        x = self.transformer.ln_f(x)
        if targets is not None:
            logits = self.lm_head(x)
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))
        else:
            logits = self.lm_head(x[:, [-1], :])
            loss = None
        return logits, loss

    @classmethod
    def from_pretrained(cls, model_type):
        assert model_type in {'gpt2', 'gpt-medium', 'gpt2-large', 'gpt2-xl'}
        from transformers import GPT2LMHeadModel
        print("loading weights from pretrained gpt: %s" % model_type)

        config_args = {
            'gpt2':       dict(n_layer=12, n_head=12, n_embd=768),
            'gpt-medium': dict(n_layer=24, n_head=16, n_embd=1024),
            'gpt2-large': dict(n_layer=36, n_head=20, n_embd=1280),
            'gpt2-xl':    dict(n_layer=48, n_head=25, n_embd=1600),
        }[model_type]
        config_args['vocab_size'] = 50257
        config_args['block_size'] = 512
        config_args['bias'] = True
        config = GPTConfig(**config_args)
        model = GPT(config)
        sd = model.state_dict()
        sd_keys = [k for k in sd.keys() if not k.endswith('.attn.bias')]

        model_hf = GPT2LMHeadModel.from_pretrained(model_type)
        sd_hf = model_hf.state_dict()
        sd_keys_hf = [k for k in sd_hf.keys() if not k.endswith('.attn.masked_bias') and not k.endswith('.attn.bias')]
        transposed = ['attn.c_attn.weight', 'attn.c_proj.weight', 'mlp.c_fc.weight', 'mlp.c_proj.weight']
        assert len(sd_keys_hf) == len(sd_keys), f"mismatched keys: {len(sd_keys_hf)} != {len(sd_keys)}"
        for k in sd_keys_hf:
            if any(k.endswith(w) for w in transposed):
                assert sd_hf[k].shape[::-1] == sd[k].shape
                with torch.no_grad():
                    sd[k].copy_(sd_hf[k].t())
            else:
                assert sd_hf[k].shape == sd[k].shape
                with torch.no_grad():
                    sd[k].copy_(sd_hf[k])
        return model

    def configure_optimizers(self, weight_decay, learning_rate, betas, device_type):
        param_dict = {pn: p for pn, p in self.named_parameters() if p.requires_grad}
        decay_params = [p for n, p in param_dict.items() if p.dim() >= 2]
        nodecay_params = [p for n, p in param_dict.items() if p.dim() < 2]
        optim_groups = [
            {'params': decay_params, 'weight_decay': weight_decay},
            {'params': nodecay_params, 'weight_decay': 0.0}
        ]
        num_decay_params = sum(p.numel() for p in decay_params)
        num_nodecay_params = sum(p.numel() for p in nodecay_params)
        print(f"num decayed parameter tensors: {len(decay_params)}, with {num_decay_params:,} parameters")
        print(f"num non-decayed parameter tensors: {len(nodecay_params)}, with {num_nodecay_params:,} parameters")
        fused_available = 'fused' in inspect.signature(torch.optim.AdamW).parameters
        use_fused = fused_available and device_type == 'cuda'
        print(f"using fused AdamW: {use_fused}")
        optimizer = torch.optim.AdamW(optim_groups, lr=learning_rate, betas=betas, eps=1e-8, fused=use_fused)
        return optimizer
```

```python
import math
import torch
import torch.nn as nn
import tiktoken
import inspect
from dataclasses import dataclass
from torch.nn import functional as F
from google.colab import drive

class LayerNorm_QKNORM(nn.Module):
    """ LayerNorm but with an optional bias. PyTorch doesn't support simply bias=False """
    def __init__(self, ndim, bias):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(ndim))
        self.bias = nn.Parameter(torch.zeros(ndim)) if bias else None

    def forward(self, input):
        return F.layer_norm(input, self.weight.shape, self.weight, self.bias, 1e-5)

class CausalSelfAttention_QKNORM(nn.Module):
    def __init__(self, config):
        super().__init__()
        assert config.n_embd % config.n_head == 0
        self.c_attn = nn.Linear(config.n_embd, 3 * config.n_embd, bias=config.bias)
        self.c_proj = nn.Linear(config.n_embd, config.n_embd, bias=config.bias)
        self.attn_dropout = nn.Dropout(config.dropout)
        self.resid_dropout = nn.Dropout(config.dropout)
        self.n_head = config.n_head
        self.n_embd = config.n_embd
        self.dropout = config.dropout
        self.flash = hasattr(torch.nn.functional, 'scaled_dot_product_attention')
        if not self.flash:
            print("WARNING: using slow attention. Flash Attention requires PyTorch >= 2.0")
            self.register_buffer("bias", torch.tril(torch.ones(config.block_size, config.block_size))
                                        .view(1, 1, config.block_size, config.block_size))
        # Add layer normalization for q and k
        self.q_norm = LayerNorm_QKNORM(config.n_embd // config.n_head, bias=config.bias)
        self.k_norm = LayerNorm_QKNORM(config.n_embd // config.n_head, bias=config.bias)

    def forward(self, x):
        B, T, C = x.size()
        qkv = self.c_attn(x)
        q, k, v = qkv.split(self.n_embd, dim=2)
        k = k.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)
        q = q.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)
        v = v.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)

        # Apply normalization to q and k
        q = self.q_norm(q)
        k = self.k_norm(k)

        if self.flash:
            y = torch.nn.functional.scaled_dot_product_attention(q, k, v, attn_mask=None, dropout_p=self.dropout if self.training else 0, is_causal=True)
        else:
            att = (q @ k.transpose(-2, -1)) * (1.0 / math.sqrt(k.size(-1)))
            att = att.masked_fill(self.bias[:,:,:T,:T] == 0, float('-inf'))
            att = F.softmax(att, dim=-1)
            att = self.attn_dropout(att)
            y = att @ v
        y = y.transpose(1, 2).contiguous().view(B, T, C)
        y = self.resid_dropout(self.c_proj(y))
        return y

class MLP_QKNORM(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.c_fc   = nn.Linear(config.n_embd, 4 * config.n_embd, bias=config.bias)
        self.gelu   = nn.GELU(approximate='tanh')
        self.c_proj = nn.Linear(4 * config.n_embd, config.n_embd, bias=config.bias)
        self.dropout = nn.Dropout(config.dropout)

    def forward(self, x):
        x = self.c_fc(x)
        x = self.gelu(x)
        x = self.c_proj(x)
        x = self.dropout(x)
        return x

class Block_QKNORM(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.ln_1 = LayerNorm_QKNORM(config.n_embd, bias=config.bias)
        self.attn = CausalSelfAttention_QKNORM(config)
        self.ln_2 = LayerNorm_QKNORM(config.n_embd, bias=config.bias)
        self.mlp = MLP_QKNORM(config)

    def forward(self, x):
        x = x + self.attn(self.ln_1(x))
        x = x + self.mlp(self.ln_2(x))
        return x

@dataclass
class GPTConfig_QKNORM:
    block_size: int = 1024
    vocab_size: int = 50304
    n_layer: int = 12
    n_head: int = 12
    n_embd: int = 768
    dropout: float = 0.0
    bias: bool = True

class GPT_QKNORM(nn.Module):
    def __init__(self, config):
        super().__init__()
        assert config.vocab_size is not None
        assert config.block_size is not None
        self.config = config

        self.transformer = nn.ModuleDict(dict(
            wte = nn.Embedding(config.vocab_size, config.n_embd),
            drop = nn.Dropout(config.dropout),
            h = nn.ModuleList([Block_QKNORM(config) for _ in range(config.n_layer)]),
            ln_f = LayerNorm_QKNORM(config.n_embd, bias=config.bias),
        ))
        self.lm_head = nn.Linear(config.n_embd, config.vocab_size, bias=False)
        self.transformer.wte.weight = self.lm_head.weight

        self.apply(self._init_weights)
        for pn, p in self.named_parameters():
            if pn.endswith('c_proj.weight'):
                torch.nn.init.normal_(p, mean=0.0, std=0.02/math.sqrt(2 * config.n_layer))

        print("number of parameters: %.2fM" % (self.get_num_params()/1e6,))

    def get_num_params(self, non_embedding=True):
        n_params = sum(p.numel() for p in self.parameters())
        return n_params

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                torch.nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, idx, targets=None):
        device = idx.device
        B, T = idx.size()
        assert T <= self.config.block_size, f"Cannot forward sequence of length {T}, block size is only {self.config.block_size}"

        tok_emb = self.transformer.wte(idx)  # (B, T, n_embd)
        x = self.transformer.drop(tok_emb)

        for block in self.transformer.h:
            x = block(x)
        x = self.transformer.ln_f(x)
        if targets is not None:
            logits = self.lm_head(x)
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))
        else:
            logits = self.lm_head(x[:, [-1], :])
            loss = None
        return logits, loss

    @classmethod
    def from_pretrained(cls, model_type):
        assert model_type in {'gpt2', 'gpt-medium', 'gpt2-large', 'gpt2-xl'}
        from transformers import GPT2LMHeadModel
        print("loading weights from pretrained gpt: %s" % model_type)

        config_args = {
            'gpt2':       dict(n_layer=12, n_head=12, n_embd=768),
            'gpt-medium': dict(n_layer=24, n_head=16, n_embd=1024),
            'gpt2-large': dict(n_layer=36, n_head=20, n_embd=1280),
            'gpt2-xl':    dict(n_layer=48, n_head=25, n_embd=1600),
        }[model_type]
        config_args['vocab_size'] = 50257
        config_args['block_size'] = 512
        config_args['bias'] = True
        config = GPTConfig(**config_args)
        model = GPT(config)
        sd = model.state_dict()
        sd_keys = [k for k in sd.keys() if not k.endswith('.attn.bias')]

        model_hf = GPT2LMHeadModel.from_pretrained(model_type)
        sd_hf = model_hf.state_dict()
        sd_keys_hf = [k for k in sd_hf.keys() if not k.endswith('.attn.masked_bias') and not k.endswith('.attn.bias')]
        transposed = ['attn.c_attn.weight', 'attn.c_proj.weight', 'mlp.c_fc.weight', 'mlp.c_proj.weight']
        assert len(sd_keys_hf) == len(sd_keys), f"mismatched keys: {len(sd_keys_hf)} != {len(sd_keys)}"
        for k in sd_keys_hf:
            if any(k.endswith(w) for w in transposed):
                assert sd_hf[k].shape[::-1] == sd[k].shape
                with torch.no_grad():
                    sd[k].copy_(sd_hf[k].t())
            else:
                assert sd_hf[k].shape == sd[k].shape
                with torch.no_grad():
                    sd[k].copy_(sd_hf[k])
        return model

    def configure_optimizers(self, weight_decay, learning_rate, betas, device_type):
        param_dict = {pn: p for pn, p in self.named_parameters() if p.requires_grad}
        decay_params = [p for n, p in param_dict.items() if p.dim() >= 2]
        nodecay_params = [p for n, p in param_dict.items() if p.dim() < 2]
        optim_groups = [
            {'params': decay_params, 'weight_decay': weight_decay},
            {'params': nodecay_params, 'weight_decay': 0.0}
        ]
        num_decay_params = sum(p.numel() for p in decay_params)
        num_nodecay_params = sum(p.numel() for p in nodecay_params)
        print(f"num decayed parameter tensors: {len(decay_params)}, with {num_decay_params:,} parameters")
        print(f"num non-decayed parameter tensors: {len(nodecay_params)}, with {num_nodecay_params:,} parameters")
        fused_available = 'fused' in inspect.signature(torch.optim.AdamW).parameters
        use_fused = fused_available and device_type == 'cuda'
        print(f"using fused AdamW: {use_fused}")
        optimizer = torch.optim.AdamW(optim_groups, lr=learning_rate, betas=betas, eps=1e-8, fused=use_fused)
        return optimizer
```

# Using RNoPE

```python
import math
import torch
import torch.nn as nn
import tiktoken
import inspect
from dataclasses import dataclass
from torch.nn import functional as F
from google.colab import drive

class LayerNorm_RNoPE(nn.Module):
    """LayerNorm with optional bias"""
    def __init__(self, ndim, bias):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(ndim))
        self.bias = nn.Parameter(torch.zeros(ndim)) if bias else None

    def forward(self, input):
        return F.layer_norm(input, self.weight.shape, self.weight, self.bias, 1e-5)

# Rotary Embedding Function
def apply_rope(q, k):
    """Apply rotary embeddings to query and key tensors"""
    B, nh, T, hs = q.shape
    device = q.device
    # Compute rotary frequencies: theta_i = 10000^(-2i/hs) for i=0 to hs//2-1
    theta = 10000 ** (-2 * torch.arange(0, hs//2).float() / hs).to(device)
    m = torch.arange(T, device=device).unsqueeze(1)  # (T, 1)
    angles = m * theta  # (T, hs//2)
    cos_vals = torch.cos(angles).unsqueeze(0).unsqueeze(0)  # (1, 1, T, hs//2)
    sin_vals = torch.sin(angles).unsqueeze(0).unsqueeze(0)  # (1, 1, T, hs//2)
    # Reshape q and k to (B, nh, T, hs//2, 2) for rotation
    q_ = q.view(B, nh, T, hs//2, 2)
    k_ = k.view(B, nh, T, hs//2, 2)
    # Apply rotation to each pair of dimensions
    q_rot = torch.stack([
        q_[..., 0] * cos_vals - q_[..., 1] * sin_vals,
        q_[..., 0] * sin_vals + q_[..., 1] * cos_vals
    ], dim=-1).view(B, nh, T, hs)
    k_rot = torch.stack([
        k_[..., 0] * cos_vals - k_[..., 1] * sin_vals,
        k_[..., 0] * sin_vals + k_[..., 1] * cos_vals
    ], dim=-1).view(B, nh, T, hs)
    return q_rot, k_rot

# Modified CausalSelfAttention with optional RNoPE
class CausalSelfAttention_RNoPE(nn.Module):
    def __init__(self, config, apply_rope=False):
        super().__init__()
        assert config.n_embd % config.n_head == 0
        self.c_attn = nn.Linear(config.n_embd, 3 * config.n_embd, bias=config.bias)
        self.c_proj = nn.Linear(config.n_embd, config.n_embd, bias=config.bias)
        self.attn_dropout = nn.Dropout(config.dropout)
        self.resid_dropout = nn.Dropout(config.dropout)
        self.n_head = config.n_head
        self.n_embd = config.n_embd
        self.dropout = config.dropout
        self.apply_rope = apply_rope  # Flag to apply rotary embeddings
        self.flash = hasattr(torch.nn.functional, 'scaled_dot_product_attention')
        if not self.flash:
            print("WARNING: using slow attention. Flash Attention requires PyTorch >= 2.0")
            self.register_buffer("bias", torch.tril(torch.ones(config.block_size, config.block_size))
                                        .view(1, 1, config.block_size, config.block_size))

    def forward(self, x):
        B, T, C = x.size()
        qkv = self.c_attn(x)
        q, k, v = qkv.split(self.n_embd, dim=2)
        k = k.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)
        q = q.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)
        v = v.view(B, T, self.n_head, C // self.n_head).transpose(1, 2)  # (B, nh, T, hs)

        # Apply rotary embeddings if enabled
        if self.apply_rope:
            q, k = apply_rope(q, k)

        if self.flash:
            y = torch.nn.functional.scaled_dot_product_attention(q, k, v, attn_mask=None, dropout_p=self.dropout if self.training else 0, is_causal=True)
        else:
            att = (q @ k.transpose(-2, -1)) * (1.0 / math.sqrt(k.size(-1)))
            att = att.masked_fill(self.bias[:,:,:T,:T] == 0, float('-inf'))
            att = F.softmax(att, dim=-1)
            att = self.attn_dropout(att)
            y = att @ v
        y = y.transpose(1, 2).contiguous().view(B, T, C)
        y = self.resid_dropout(self.c_proj(y))
        return y

# FeedForward Network (unchanged)
class MLP_RNoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.c_fc = nn.Linear(config.n_embd, 4 * config.n_embd, bias=config.bias)
        self.gelu = nn.GELU(approximate='tanh')
        self.c_proj = nn.Linear(4 * config.n_embd, config.n_embd, bias=config.bias)
        self.dropout = nn.Dropout(config.dropout)

    def forward(self, x):
        x = self.c_fc(x)
        x = self.gelu(x)
        x = self.c_proj(x)
        x = self.dropout(x)
        return x

# Modified Transformer Block with RNoPE option
class Block_RNoPE(nn.Module):
    def __init__(self, config, apply_rope=False):
        super().__init__()
        self.ln_1 = LayerNorm_RNoPE(config.n_embd, bias=config.bias)
        self.attn = CausalSelfAttention_RNoPE(config, apply_rope=apply_rope)
        self.ln_2 = LayerNorm_RNoPE(config.n_embd, bias=config.bias)
        self.mlp = MLP_RNoPE(config)

    def forward(self, x):
        x = x + self.attn(self.ln_1(x))
        x = x + self.mlp(self.ln_2(x))
        return x

# GPT Configuration (unchanged)
@dataclass
class GPTConfig_RNoPE:
    block_size: int = 1024
    vocab_size: int = 50304
    n_layer: int = 12
    n_head: int = 12
    n_embd: int = 768
    dropout: float = 0.0
    bias: bool = True

# Modified GPT Model with RNoPE
class GPT_RNoPE(nn.Module):
    def __init__(self, config):
        super().__init__()
        assert config.vocab_size is not None
        assert config.block_size is not None
        self.config = config

        self.transformer = nn.ModuleDict(dict(
            wte = nn.Embedding(config.vocab_size, config.n_embd),
            drop = nn.Dropout(config.dropout),
            h = nn.ModuleList([Block_RNoPE(config, apply_rope=(i % 2 == 0)) for i in range(config.n_layer)]),  # Alternating RoPE
            ln_f = LayerNorm_RNoPE(config.n_embd, bias=config.bias),
        ))
        self.lm_head = nn.Linear(config.n_embd, config.vocab_size, bias=False)
        self.transformer.wte.weight = self.lm_head.weight  # Weight tying

        self.apply(self._init_weights)
        for pn, p in self.named_parameters():
            if pn.endswith('c_proj.weight'):
                torch.nn.init.normal_(p, mean=0.0, std=0.02/math.sqrt(2 * config.n_layer))

        print("number of parameters: %.2fM" % (self.get_num_params()/1e6,))

    def get_num_params(self, non_embedding=True):
        """Return total number of parameters; no positional embeddings to subtract"""
        n_params = sum(p.numel() for p in self.parameters())
        return n_params

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                torch.nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, idx, targets=None):
        device = idx.device
        B, T = idx.size()
        assert T <= self.config.block_size, f"Cannot forward sequence of length {T}, block size is only {self.config.block_size}"

        # Only token embeddings (no positional embeddings)
        tok_emb = self.transformer.wte(idx)  # (B, T, n_embd)
        x = self.transformer.drop(tok_emb)
        for block in self.transformer.h:
            x = block(x)
        x = self.transformer.ln_f(x)
        if targets is not None:
            logits = self.lm_head(x)
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))
        else:
            logits = self.lm_head(x[:, [-1], :])
            loss = None
        return logits, loss

    @classmethod
    def from_pretrained(cls, model_type):
        assert model_type in {'gpt2', 'gpt-medium', 'gpt2-large', 'gpt2-xl'}
        from transformers import GPT2LMHeadModel
        print("loading weights from pretrained gpt: %s" % model_type)

        config_args = {
            'gpt2':       dict(n_layer=12, n_head=12, n_embd=768),
            'gpt-medium': dict(n_layer=24, n_head=16, n_embd=1024),
            'gpt2-large': dict(n_layer=36, n_head=20, n_embd=1280),
            'gpt2-xl':    dict(n_layer=48, n_head=25, n_embd=1600),
        }[model_type]
        config_args['vocab_size'] = 50257
        config_args['block_size'] = 512
        config_args['bias'] = True
        config = GPTConfig(**config_args)
        model = GPT(config)
        sd = model.state_dict()
        sd_keys = [k for k in sd.keys() if not k.endswith('.attn.bias')]

        model_hf = GPT2LMHeadModel.from_pretrained(model_type)
        sd_hf = model_hf.state_dict()
        sd_keys_hf = [k for k in sd_hf.keys() if not k.endswith('.attn.masked_bias') and not k.endswith('.attn.bias')]
        transposed = ['attn.c_attn.weight', 'attn.c_proj.weight', 'mlp.c_fc.weight', 'mlp.c_proj.weight']

        # Copy only matching keys, skipping 'transformer.wpe.weight'
        for k in sd_keys_hf:
            if k in sd:
                if any(k.endswith(w) for w in transposed):
                    assert sd_hf[k].shape[::-1] == sd[k].shape
                    with torch.no_grad():
                        sd[k].copy_(sd_hf[k].t())
                else:
                    assert sd_hf[k].shape == sd[k].shape
                    with torch.no_grad():
                        sd[k].copy_(sd_hf[k])
        return model

    def configure_optimizers(self, weight_decay, learning_rate, betas, device_type):
        param_dict = {pn: p for pn, p in self.named_parameters() if p.requires_grad}
        decay_params = [p for n, p in param_dict.items() if p.dim() >= 2]
        nodecay_params = [p for n, p in param_dict.items() if p.dim() < 2]
        optim_groups = [
            {'params': decay_params, 'weight_decay': weight_decay},
            {'params': nodecay_params, 'weight_decay': 0.0}
        ]
        num_decay_params = sum(p.numel() for p in decay_params)
        num_nodecay_params = sum(p.numel() for p in nodecay_params)
        print(f"num decayed parameter tensors: {len(decay_params)}, with {num_decay_params:,} parameters")
        print(f"num non-decayed parameter tensors: {len(nodecay_params)}, with {num_nodecay_params:,} parameters")
        fused_available = 'fused' in inspect.signature(torch.optim.AdamW).parameters
        use_fused = fused_available and device_type == 'cuda'
        print(f"using fused AdamW: {use_fused}")
        optimizer = torch.optim.AdamW(optim_groups, lr=learning_rate, betas=betas, eps=1e-8, fused=use_fused)
        return optimizer
```

# Pretraining


```python
import os
import time
import math
import pickle
from contextlib import nullcontext
import numpy as np
import torch
from torch.amp import autocast, GradScaler
from tqdm import tqdm  # Ensure tqdm is installed: pip install tqdm

# Global configurations
out_dir = 'out'
train_iters = 10000  # For longer training, increase this (e.g., 5000)
val_iters = 50    # Can reduce to 50 for faster periodic validation
val_interval = 100  # Validate every 100 iterations
log_interval = 100  # Not used in this modified code
gradient_accumulation_steps = 4
batch_size = 32
block_size = 1024
n_layer = 4
n_head = 4
n_embd = 256
dropout = 0.05
bias = False
learning_rate = 6e-4
weight_decay = 1e-3
beta1 = 0.9
beta2 = 0.95
grad_clip = 1.0
warmup_iters = 20
lr_decay_iters = 10000
min_lr = 6e-5
device = 'cuda' if torch.cuda.is_available() else 'cpu'
dtype = 'float16'
device_type = 'cuda' if 'cuda' in device else 'cpu'
ptdtype = {'float32': torch.float32, 'bfloat16': torch.bfloat16, 'float16': torch.float16}[dtype]
data_dir = os.path.join('/content/drive/MyDrive/datasets', 'input')

os.makedirs(out_dir, exist_ok=True)
torch.manual_seed(1337)
if device == 'cuda':
    torch.backends.cuda.matmul.allow_tf32 = True
    torch.backends.cudnn.allow_tf32 = True
ctx = nullcontext() if device_type == 'cpu' else autocast(device_type=device_type, dtype=ptdtype)

# Data loading function
def get_batch(split):
    data_file = os.path.join(data_dir, 'train.bin' if split == 'train' else 'val.bin')
    data = np.memmap(data_file, dtype=np.uint16, mode='r')
    ix = torch.randint(0, len(data) - block_size, (batch_size,))
    x = torch.stack([torch.from_numpy(data[i:i+block_size].astype(np.int64)) for i in ix])
    y = torch.stack([torch.from_numpy(data[i+1:i+1+block_size].astype(np.int64)) for i in ix])
    return x.pin_memory().to(device, non_blocking=True), y.pin_memory().to(device, non_blocking=True)

# Learning rate scheduler
def get_lr(it):
    if it < warmup_iters:
        return learning_rate * (it + 1) / warmup_iters
    if it > lr_decay_iters:
        return min_lr
    decay_ratio = (it - warmup_iters) / (lr_decay_iters - warmup_iters)
    return min_lr + 0.5 * (learning_rate - min_lr) * (1 + math.cos(math.pi * decay_ratio))

# Training and validation function
def train_and_validate(model_class, config_class, model_args, pe):
    """
    Trains and validates a model, printing losses at the start and every validation interval.

    Args:
        model_class: The model class (e.g., GPT, NoPE_GPT)
        config_class: The config class (e.g., GPTConfig, NoPE_GPTConfig)
        model_args: Dictionary of model hyperparameters
        pe: String identifier for positional encoding (e.g., 'absolute_pe')

    Returns:
        tuple: (train_avg_losses, val_losses) - lists of average training and validation losses
               at step 0 and every val_interval iterations
    """
    # Load vocab size from meta file, or use default
    meta_path = os.path.join(data_dir, 'meta.pkl')
    if os.path.exists(meta_path):
        with open(meta_path, 'rb') as f:
            meta = pickle.load(f)
        model_args['vocab_size'] = meta['vocab_size']
    else:
        model_args['vocab_size'] = 50304

    # Initialize model
    config = config_class(**model_args)
    model = model_class(config)
    model.to(device)

    # Compile the model if torch.compile is available
    if hasattr(torch, 'compile'):
        model = torch.compile(model)
        print("Model compiled with torch.compile for potential performance improvement.")
    else:
        print("torch.compile is not available; proceeding without compilation.")

    # Optimizer and scaler
    optimizer = model.configure_optimizers(weight_decay, learning_rate, (beta1, beta2), device_type)
    scaler = GradScaler(enabled=(dtype == 'float16'))

    # Compute initial training loss (in train mode)
    model.train()
    x, y = get_batch('train')
    with ctx:
        logits, loss = model(x, y)
    initial_train_loss = loss.item()

    # Compute initial validation loss (in eval mode)
    model.eval()
    val_loss_sum = 0.0
    with torch.no_grad():
        for _ in range(val_iters):
            x_val, y_val = get_batch('val')
            with ctx:
                logits, loss = model(x_val, y_val)
            val_loss_sum += loss.item()
    initial_val_loss = val_loss_sum / val_iters

    # Print initial losses as Step 0
    step_counter = 0
    print(f"Step {step_counter} | GPT with {pe}: Train Loss: {initial_train_loss:.4f} | Val Loss: {initial_val_loss:.4f}")

    # Initialize loss lists with initial values
    train_avg_losses = [initial_train_loss]
    val_losses = [initial_val_loss]

    # Training loop
    model.train()
    train_loss_sum = 0.0
    x, y = get_batch('train')
    for iter in tqdm(range(train_iters), desc=f"Training {pe}", unit="iter"):
        lr = get_lr(iter)
        for param_group in optimizer.param_groups:
            param_group['lr'] = lr
        optimizer.zero_grad()
        loss_accum = 0.0

        for micro_step in range(gradient_accumulation_steps):
            with ctx:
                logits, loss = model(x, y)
                loss = loss / gradient_accumulation_steps
            # Update the batch for the next micro step
            x, y = get_batch('train')
            loss_accum += loss.item()
            scaler.scale(loss).backward()
        if grad_clip:
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
        scaler.step(optimizer)
        scaler.update()

        train_loss_sum += loss_accum

        # Every val_interval iterations or at the end, compute and print averaged losses
        if (iter + 1) % val_interval == 0 or (iter + 1) == train_iters:
            steps_in_interval = val_interval if (iter + 1) % val_interval == 0 else (iter + 1) % val_interval
            avg_train_loss = train_loss_sum / steps_in_interval
            train_avg_losses.append(avg_train_loss)

            model.eval()
            val_loss_sum = 0.0
            with torch.no_grad():
                for _ in range(val_iters):
                    x_val, y_val = get_batch('val')
                    with ctx:
                        logits, loss = model(x_val, y_val)
                    val_loss_sum += loss.item()
            avg_val_loss = val_loss_sum / val_iters
            val_losses.append(avg_val_loss)

            # Increment the step counter and print summary loss
            step_counter += 1
            print(f"\nStep {iter + 1} | GPT with {pe}: Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")
            train_loss_sum = 0.0
            model.train()

    # Release GPU memory
    del model
    del optimizer
    if device == 'cuda':
        torch.cuda.empty_cache()

    return train_avg_losses, val_losses
```

```python
# Model arguments
model_args = dict(
    n_layer=n_layer,
    n_head=n_head,
    n_embd=n_embd,
    block_size=block_size,
    bias=bias,
    dropout=dropout
)

# Initialize an empty dictionary to store all losses
all_losses = {}
```

```python
'''
Naive Implementation
class MLA(nn.Module):
    """
    Multi-Headed Attention Layer (MLA) from DeepSeek.
    """
    def __init__(self, args):
        super().__init__()
        self.dim = args.n_embd
        self.n_heads = args.n_head
        self.n_local_heads = args.n_head  # No distributed processing for simplicity

        # LoRA ranks (set to default values if not in args)
        self.q_lora_rank = getattr(args, 'q_lora_rank', 0)
        self.kv_lora_rank = getattr(args, 'kv_lora_rank', 64)

        # Head dimensions
        head_dim = self.dim // self.n_heads
        self.qk_rope_head_dim = getattr(args, 'qk_rope_head_dim', head_dim // 2)
        self.qk_nope_head_dim = head_dim - self.qk_rope_head_dim
        self.qk_head_dim = head_dim
        self.v_head_dim = head_dim

        # Scaling factor
        self.softmax_scale = self.qk_head_dim ** -0.5

        # Query projection with optional LoRA
        if self.q_lora_rank == 0:
            self.wq = nn.Linear(self.dim, self.n_heads * self.qk_head_dim, bias=args.bias)
        else:
            self.wq_a = nn.Linear(self.dim, self.q_lora_rank, bias=args.bias)
            self.q_norm = RMSNorm(self.q_lora_rank)
            self.wq_b = nn.Linear(self.q_lora_rank, self.n_heads * self.qk_head_dim, bias=args.bias)

        # Key-value projection with LoRA
        self.wkv_a = nn.Linear(self.dim, self.kv_lora_rank + self.qk_rope_head_dim, bias=args.bias)
        self.kv_norm = RMSNorm(self.kv_lora_rank)
        self.wkv_b = nn.Linear(self.kv_lora_rank, self.n_heads * (self.qk_nope_head_dim + self.v_head_dim), bias=args.bias)

        # Output projection
        self.wo = nn.Linear(self.n_heads * self.v_head_dim, self.dim, bias=args.bias)

        # Setup caching for inference
        self.max_batch_size = getattr(args, 'max_batch_size', 8)
        self.max_seq_len = getattr(args, 'block_size', 1024)

        # Register buffers for KV caching
        self.register_buffer("k_cache", torch.zeros(
            self.max_batch_size, self.max_seq_len, self.n_local_heads, self.qk_head_dim
        ), persistent=False)
        self.register_buffer("v_cache", torch.zeros(
            self.max_batch_size, self.max_seq_len, self.n_local_heads, self.v_head_dim
        ), persistent=False)

        # For causal masking in non-flash attention mode
        self.register_buffer("causal_mask", torch.tril(torch.ones(self.max_seq_len, self.max_seq_len))
                                           .view(1, 1, self.max_seq_len, self.max_seq_len))

        # Flash attention support
        self.flash = hasattr(torch.nn.functional, 'scaled_dot_product_attention')

    def forward(self, x, start_pos, freqs_cis, mask=None):
        """
        Forward pass for the attention layer.

        Args:
            x (torch.Tensor): Input tensor of shape (batch_size, seq_len, dim).
            start_pos (int): Starting position in the sequence for caching.
            freqs_cis (torch.Tensor): Precomputed complex exponential values for rotary embeddings.
            mask (Optional[torch.Tensor]): Mask tensor to exclude certain positions from attention.

        Returns:
            torch.Tensor: Output tensor after attention.
        """
        bsz, seqlen, _ = x.size()
        end_pos = start_pos + seqlen

        # Project queries
        if self.q_lora_rank == 0:
            q = self.wq(x)
        else:
            q = self.wq_b(self.q_norm(self.wq_a(x)))

        q = q.view(bsz, seqlen, self.n_local_heads, self.qk_head_dim)

        # Split into rotary and non-rotary parts
        q_nope, q_pe = torch.split(q, [self.qk_nope_head_dim, self.qk_rope_head_dim], dim=-1)

        # Apply rotary embeddings to q_pe
        q_pe = apply_rotary_emb(q_pe, freqs_cis)

        # Project key/value with LoRA
        kv = self.wkv_a(x)
        kv, k_pe = torch.split(kv, [self.kv_lora_rank, self.qk_rope_head_dim], dim=-1)

        # Apply rotary embeddings to k_pe
        k_pe = apply_rotary_emb(k_pe.unsqueeze(2), freqs_cis)

        # Convert low-rank representation to key/value
        kv = self.wkv_b(self.kv_norm(kv))
        kv = kv.view(bsz, seqlen, self.n_local_heads, self.qk_nope_head_dim + self.v_head_dim)
        k_nope, v = torch.split(kv, [self.qk_nope_head_dim, self.v_head_dim], dim=-1)

        # Combine rotary and non-rotary parts
        k = torch.cat([k_nope, k_pe.expand(-1, -1, self.n_local_heads, -1)], dim=-1)

        # Update caches
        self.k_cache[:bsz, start_pos:end_pos] = k.detach()
        self.v_cache[:bsz, start_pos:end_pos] = v.detach()

        # Recombine q for attention computation
        q = torch.cat([q_nope, q_pe], dim=-1)

        # Transpose for attention computation: (B, T, nh, hd) -> (B, nh, T, hd)
        q = q.transpose(1, 2)

        if self.flash:
            # Use Flash Attention if available
            k_cache = self.k_cache[:bsz, :end_pos].transpose(1, 2)
            v_cache = self.v_cache[:bsz, :end_pos].transpose(1, 2)

            y = torch.nn.functional.scaled_dot_product_attention(
                q, k_cache, v_cache,
                attn_mask=None,
                dropout_p=0.0,
                is_causal=True
            )
        else:
            # Manual implementation using cached keys and values
            k_cache = self.k_cache[:bsz, :end_pos].transpose(1, 2)
            v_cache = self.v_cache[:bsz, :end_pos].transpose(1, 2)

            scores = torch.matmul(q, k_cache.transpose(-2, -1)) * self.softmax_scale

            # Apply causal mask
            if mask is None:
                mask_t = self.causal_mask[:, :, :q.size(2), :end_pos]
                scores = scores.masked_fill(mask_t == 0, float('-inf'))
            else:
                scores = scores + mask.unsqueeze(1)

            scores = F.softmax(scores, dim=-1, dtype=torch.float32).type_as(x)
            y = torch.matmul(scores, v_cache)

        # Reshape output: (B, nh, T, hd) -> (B, T, nh*hd)
        y = y.transpose(1, 2).contiguous().view(bsz, seqlen, -1)

        # Output projection
        y = self.wo(y)

        return y
'''
```

```python
import math
import torch
import torch.nn as nn
import torch.nn.functional as F
from typing import Optional, Tuple


class Linear(nn.Module):
    def __init__(self, in_features: int, out_features: int, bias: bool = False, dtype=torch.float16):
        super().__init__()
        self.in_features = in_features
        self.out_features = out_features

        # Main weight tensor
        self.weight = nn.Parameter(torch.empty(out_features, in_features, dtype=dtype))

        # GPT-2 style initialization scaling
        self._init_weights()

        # Quantization parameters
        self.quantized = False
        if dtype in [torch.int8, torch.uint8]:  # Add quantization if needed
            self._init_quantization()

        # Bias handling
        if bias:
            self.bias = nn.Parameter(torch.empty(out_features))
            nn.init.zeros_(self.bias)
        else:
            self.register_parameter('bias', None)

    def _init_weights(self):
        std = 0.02
        if hasattr(self, 'NANOGPT_SCALE_INIT'):
            std *= (2 * self.config.n_layer) ** -0.5
        nn.init.normal_(self.weight, mean=0.0, std=std)

    def _init_quantization(self):
        """Initialize quantization parameters"""
        self.quantized = True
        self.scale = nn.Parameter(torch.ones(
            (self.out_features + block_size - 1) // block_size,
            (self.in_features + block_size - 1) // block_size,
            dtype=torch.float32
        ))

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # Main computation path
        if not self.quantized or gemm_impl == "f16":
            return F.linear(x, self.weight, self.bias)

        # Quantized path
        x_q, x_scale = act_quant(x, block_size)
        w_deq = weight_dequant(self.weight, self.scale)
        return fp8_gemm(x_q, x_scale, w_deq, self.scale, self.bias)

def linear(x: torch.Tensor, weight: torch.Tensor, bias: Optional[torch.Tensor] = None) -> torch.Tensor:
    if weight.requires_grad:
        # Use standard linear for trainable weights
        return F.linear(x, weight, bias)

    # Quantization handling for inference
    if hasattr(weight, 'scale'):
        return fp8_gemm(
            act_quant(x, block_size)[0],
            weight_dequant(weight, weight.scale),
            bias
        )
    return F.linear(x, weight, bias)

class ColumnParallelLinear(Linear):

    def __init__(self, in_features: int, out_features: int, bias: bool = False, dtype = None):
        assert out_features % world_size == 0, (
            f"Output features ({out_features}) must be divisible by world size ({world_size})"
        )
        self.part_out_features = out_features // world_size
        super().__init__(in_features, self.part_out_features, bias, dtype)

        # GPT-2 style initialization scaling
        if hasattr(self, 'NANOGPT_SCALE_INIT'):
            std = 0.02 * (2 * self.config.n_layer) ** -0.5  # Add this if using with GPT blocks
            nn.init.normal_(self.weight, mean=0.0, std=std)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        y = super().forward(x)

        # Maintain compatibility with standard Linear output shape
        if world_size == 1:
            return y  # No change needed for single process
        return y  # Output is already sharded across columns


class RowParallelLinear(Linear):

    def __init__(self, in_features: int, out_features: int, bias: bool = False, dtype = None):
        assert in_features % world_size == 0, (
            f"Input features ({in_features}) must be divisible by world size ({world_size})"
        )
        self.part_in_features = in_features // world_size
        super().__init__(self.part_in_features, out_features, bias, dtype)

        # GPT-2 style initialization
        nn.init.normal_(self.weight, mean=0.0, std=0.02)
        if bias:
            nn.init.zeros_(self.bias)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        y = super().forward(x)

        # GPT-2 compatibility enhancements
        if world_size > 1:
            # Use mean reduction instead of sum for better gradient stability
            dist.all_reduce(y, op=dist.ReduceOp.AVG)

        # Ensure bias is applied after reduction
        if self.bias is not None:
            y += self.bias

        return y

class RMSNorm(nn.Module):
    def __init__(self, dim: int, eps: float = 1e-6):
        super().__init__()
        self.eps = eps
        self.dim = dim
        self.weight = nn.Parameter(torch.ones(dim))

    def forward(self, x):
        # Use tuple for normalized_shape and proper parameter order
        return F.rms_norm(
            x,
            normalized_shape=(self.dim,),  # Must be tuple of ints
            weight=self.weight,
            eps=self.eps
        )

def apply_rotary_emb(x, freqs_cis):
    """
    Apply rotary embeddings to the input tensor.

    Args:
        x: Input tensor of shape (batch_size, seq_len, num_heads, head_dim)
        freqs_cis: Precomputed frequency tensor of shape (seq_len, head_dim / 2)

    Returns:
        Tensor with rotary embeddings applied, same shape as input.
    """
    x_ = torch.view_as_complex(x.float().reshape(*x.shape[:-1], -1, 2))  # (batch_size, seq_len, num_heads, head_dim / 2)
    freqs_cis_ = freqs_cis.view(1, freqs_cis.size(0), 1, freqs_cis.size(1))  # (1, seq_len, 1, head_dim / 2)
    y = torch.view_as_real(x_ * freqs_cis_).flatten(3)  # Multiply and flatten
    return y.to(x.dtype)

def precompute_freqs_cis(dim: int, max_seq_len: int, theta: float = 10000.0, dtype=torch.float):
    """
    Precompute the frequency tensor for complex rotation.

    Args:
        dim: Dimension of the rotary embeddings (head_dim)
        max_seq_len: Maximum sequence length
        theta: Base value for the rotation
        dtype: Data type for the created tensors

    Returns:
        torch.Tensor: Complex rotation tensor of shape (seq_len, head_dim / 2)
    """
    # Ensure dim is even
    if dim % 2 != 0:
        raise ValueError("Dimension must be divisible by 2")

    # Compute frequencies
    freqs = 1.0 / (theta ** (torch.arange(0, dim, 2).float() / dim))
    t = torch.arange(max_seq_len, dtype=freqs.dtype)
    freqs = torch.outer(t, freqs)  # (max_seq_len, dim / 2)

    # Convert to complex exponential
    freqs_cis = torch.polar(torch.ones_like(freqs), freqs)  # (max_seq_len, dim / 2)

    return freqs_cis.to(dtype=dtype)



'''
   DeepSeek Implementation
'''
class MLA(nn.Module):
    """
    Single-GPU Optimized Multi-Headed Latent Attention
    """
    def __init__(self, args):
        super().__init__()
        self.dim = args.n_embd
        self.n_heads = args.n_head

        # LoRA configuration
        self.q_lora_rank = getattr(args, 'q_lora_rank', 0)
        self.kv_lora_rank = getattr(args, 'kv_lora_rank', 64)

        # Head dimensions configuration
        self.qk_nope_head_dim = args.qk_nope_head_dim
        self.qk_rope_head_dim = args.qk_rope_head_dim
        self.qk_head_dim = args.qk_nope_head_dim + args.qk_rope_head_dim
        self.v_head_dim = args.v_head_dim

        # Projection layers (single GPU version)
        if self.q_lora_rank == 0:
            self.wq = nn.Linear(self.dim, self.n_heads * self.qk_head_dim, bias=args.bias)
        else:
            self.wq_a = nn.Linear(self.dim, self.q_lora_rank, bias=args.bias)
            self.q_norm = RMSNorm(self.q_lora_rank)
            self.wq_b = nn.Linear(self.q_lora_rank, self.n_heads * self.qk_head_dim, bias=args.bias)

        # Key-value projections
        self.wkv_a = nn.Linear(self.dim, self.kv_lora_rank + self.qk_rope_head_dim, bias=args.bias)
        self.kv_norm = RMSNorm(self.kv_lora_rank)
        self.wkv_b = nn.Linear(self.kv_lora_rank,
                             self.n_heads * (self.qk_nope_head_dim + self.v_head_dim),
                             bias=args.bias)

        # Output projection
        self.wo = nn.Linear(self.n_heads * self.v_head_dim, self.dim, bias=args.bias)

        # Attention scaling
        self.softmax_scale = self.qk_head_dim ** -0.5
        if getattr(args, 'max_seq_len', 0) > getattr(args, 'original_seq_len', 0):
            mscale = 0.1 * args.mscale * math.log(args.rope_factor) + 1.0
            self.softmax_scale *= mscale * mscale

        # Cache configuration
        self.max_batch_size = getattr(args, 'max_batch_size', 8)
        self.max_seq_len = getattr(args, 'block_size', 1024)
        self.register_buffer("kv_cache", torch.zeros(
            self.max_batch_size, self.max_seq_len, self.kv_lora_rank
        ), persistent=False)
        self.register_buffer("pe_cache", torch.zeros(
            self.max_batch_size, self.max_seq_len, self.qk_rope_head_dim
        ), persistent=False)
        self.register_buffer("causal_mask",
                     torch.tril(torch.ones(self.max_seq_len, self.max_seq_len))
                     .view(1, self.max_seq_len, 1, self.max_seq_len),
                     persistent=False)

        # Initialize weights
        self._init_weights()

    def _init_weights(self):
        """GPT-2 style initialization"""
        std = 0.02
        nn.init.normal_(self.wq.weight, mean=0.0, std=std)
        nn.init.normal_(self.wkv_a.weight, mean=0.0, std=std)
        nn.init.normal_(self.wo.weight, mean=0.0, std=std)
        if self.q_lora_rank > 0:
            nn.init.kaiming_normal_(self.wq_a.weight, mode='fan_in', nonlinearity='linear')
            nn.init.zeros_(self.wq_b.weight)
        if hasattr(self, 'bias') and self.bias is not None:
            nn.init.zeros_(self.bias)

    def reset_cache(self):
        """Reset attention cache for inference"""
        self.kv_cache.zero_()
        self.pe_cache.zero_()

    def forward(self, x, start_pos, freqs_cis, mask=None):
        bsz, seqlen, _ = x.size()
        end_pos = start_pos + seqlen

        # Query projection
        if self.q_lora_rank == 0:
            q = self.wq(x)
        else:
            q = self.wq_b(self.q_norm(self.wq_a(x)))
        q = q.view(bsz, seqlen, self.n_heads, self.qk_head_dim)
        q_nope, q_pe = torch.split(q, [self.qk_nope_head_dim, self.qk_rope_head_dim], dim=-1)
        q_pe = apply_rotary_emb(q_pe, freqs_cis)

        # Key/Value processing
        kv = self.wkv_a(x)
        kv, k_pe = torch.split(kv, [self.kv_lora_rank, self.qk_rope_head_dim], dim=-1)
        k_pe = apply_rotary_emb(k_pe.unsqueeze(2), freqs_cis).squeeze(2)
        kv = self.kv_norm(kv)

        # Process weights
        wkv_b = self.wkv_b.weight.view(self.n_heads, -1, self.kv_lora_rank)

        # Cache handling
        if not self.training:
            self.kv_cache[:bsz, start_pos:end_pos] = kv
            self.pe_cache[:bsz, start_pos:end_pos] = k_pe
        if self.training:
            kv_combined = kv
            pe_combined = k_pe
        else:
            kv_combined = torch.cat([self.kv_cache[:bsz, :start_pos], kv], dim=1)
            pe_combined = torch.cat([self.pe_cache[:bsz, :start_pos], k_pe], dim=1)

        # Compute Q
        q_nope_transformed = torch.einsum("bshd,hdc->bshc", q_nope, wkv_b[:, :self.qk_nope_head_dim])
        Q = torch.cat([q_nope_transformed, q_pe], dim=-1).permute(0, 2, 1, 3)  # (bsz, n_heads, seqlen, kv_lora_rank + qk_rope_head_dim)

        # Compute K
        K = torch.cat([kv_combined, pe_combined], dim=-1)  # (bsz, seq_len_combined, kv_lora_rank + qk_rope_head_dim)
        K = K.unsqueeze(1).expand(-1, self.n_heads, -1, -1).permute(0, 1, 2, 3)  # (bsz, n_heads, seq_len_combined, kv_lora_rank + qk_rope_head_dim)

        # Compute V
        wkv_b_val = wkv_b[:, -self.v_head_dim:, :]  # (n_heads, v_head_dim, kv_lora_rank)
        V = torch.einsum("btc,hdc->bthd", kv_combined, wkv_b_val).permute(0, 2, 1, 3)  # (bsz, n_heads, seq_len_combined, v_head_dim)

        # Flash Attention
        with torch.backends.cuda.sdp_kernel(enable_flash=True, enable_math=False, enable_mem_efficient=False):
            x_out = torch.nn.functional.scaled_dot_product_attention(
                query=Q,
                key=K,
                value=V,
                attn_mask=mask,
                is_causal=(mask is None and seqlen > 1),
                scale=self.softmax_scale,
            )

        # Output projection
        x_out = x_out.permute(0, 2, 1, 3).reshape(bsz, seqlen, -1)  # (bsz, seqlen, n_heads * v_head_dim)
        return self.wo(x_out)

class MLP(nn.Module):
    """
    Deepseek-style MLP with model parallel components
    """
    def __init__(self, dim: int, inter_dim: int):
        super().__init__()
        # ColumnParallelLinear splits the output dimension across devices
        self.w1 = ColumnParallelLinear(
            in_features=dim,
            out_features=inter_dim,
            bias=False  # Deepseek MLP doesn't use biases
        )

        # RowParallelLinear combines the input dimension across devices
        self.w2 = RowParallelLinear(
            in_features=inter_dim,
            out_features=dim,
            bias=False
        )

        # Second ColumnParallelLinear for gating
        self.w3 = ColumnParallelLinear(
            in_features=dim,
            out_features=inter_dim,
            bias=False
        )

        # Deepseek-style initialization
        self.apply(self._init_weights)

    def _init_weights(self, module):
        print(f"Initializing {module}")
        if isinstance(module, nn.Linear):
            std = 0.02
            if hasattr(module, 'NANOGPT_SCALE_INIT'):
                std *= (2 * self.config.n_layer) ** -0.5
            elif 'wq_a' in module.__dict__ or 'wkv_a' in module.__dict__:
                nn.init.kaiming_normal_(module.weight, mode='fan_in', nonlinearity='linear')
            else:
                nn.init.normal_(module.weight, mean=0.0, std=std)
            if module.bias is not None:
                nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """
        Forward pass with model parallelism:
        1. w1 and w3 operate on split columns
        2. Element-wise multiplication happens locally
        3. w2 combines results across devices
        """
        return self.w2(F.silu(self.w1(x)) * self.w3(x))

class Block(nn.Module):
    """
    Transformer block combining attention and feed-forward layers following DeepSeek's architecture.
    """
    def __init__(self, layer_id: int, args):
        """
        Initializes the Transformer block.

        Args:
            layer_id (int): Layer index in the transformer.
            args: Model arguments containing block parameters.
        """
        super().__init__()
        self.attn = MLA(args)
        # For simplicity, we're using MLP for all layers instead of MoE
        self.ffn = MLP(args.n_embd, getattr(args, 'inter_dim', 4 * args.n_embd))
        self.attn_norm = RMSNorm(args.n_embd)
        self.ffn_norm = RMSNorm(args.n_embd)

    def forward(self, x: torch.Tensor, start_pos: int, freqs_cis: torch.Tensor, mask: Optional[torch.Tensor] = None) -> torch.Tensor:
        """
        Forward pass for the Transformer block.

        Args:
            x (torch.Tensor): Input tensor.
            start_pos (int): Starting position in the sequence.
            freqs_cis (torch.Tensor): Precomputed complex exponential values for rotary embeddings.
            mask (Optional[torch.Tensor]): Mask tensor to exclude certain positions from attention.

        Returns:
            torch.Tensor: Output tensor after block computation.
        """
        x = x + self.attn(self.attn_norm(x), start_pos, freqs_cis, mask)
        x = x + self.ffn(self.ffn_norm(x))
        return x
```

```python
import math
import torch
import torch.nn as nn
import torch.nn.functional as F
from typing import Optional, Tuple
from dataclasses import dataclass
import tiktoken
import time
from torch.amp import autocast, GradScaler

# [RMSNorm, apply_rotary_emb, precompute_freqs_cis, MLA classes remain unchanged from your provided code]

# Updated GPTConfig with bias parameter
from dataclasses import dataclass
from typing import Literal

world_size: int = 1
rank: int = 0
# Computation parameters
attn_impl: Literal["naive", "absorb"] = "absorb"  # Default to original attention
gemm_dtype: Literal["fp32", "fp16"] = "fp16"     # Match GPT-2's precision
flash_attn: bool = False      # Enable only if using flash attention
@dataclass
class GPTConfig:
    block_size: int = 1024
    vocab_size: int = 50304  # Slightly larger than GPT-2’s 50257
    n_layer: int = 12
    n_head: int = 12
    n_embd: int = 768
    bias: bool = False

    q_lora_rank: int = 0
    kv_lora_rank: int = 64
    qk_nope_head_dim: int = 64
    qk_rope_head_dim: int = 64
    v_head_dim: int = 128

    original_seq_len: int = 1024
    rope_theta: float = 10000.0
    rope_factor: float = 1.0
    mscale: float = 1.0

    max_batch_size: int = 8
    dtype: Literal["fp16", "fp32"] = "fp16"

    moe_inter_dim: int = 0
    n_routed_experts: int = 0
    n_activated_experts: int = 0

# Updated GPT model
class GPT(nn.Module):
    def __init__(self, config):
        super().__init__()
        self.config = config

        self.transformer = nn.ModuleDict(dict(
            wte = nn.Embedding(config.vocab_size, config.n_embd),
            h = nn.ModuleList([Block(i, config) for i in range(config.n_layer)]),
            ln_f = RMSNorm(config.n_embd),  # Use RMSNorm instead of LayerNorm
        ))
        self.lm_head = nn.Linear(config.n_embd, config.vocab_size, bias=False)

        # Weight sharing between token embeddings and language model head
        self.transformer.wte.weight = self.lm_head.weight

        # Precompute frequencies for rotary embeddings
        #freqs_cis = precompute_freqs_cis(config.n_embd // config.n_head, config.block_size)
        head_dim = config.n_embd // config.n_head
        # Use half of head_dim for rotary embeddings.
        freqs_cis = precompute_freqs_cis(head_dim // 2, config.block_size)
        self.register_buffer('freqs_cis', freqs_cis)


        self.register_buffer('freqs_cis', freqs_cis)

        # Initialize weights
        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            std = 0.02
            if hasattr(module, 'NANOGPT_SCALE_INIT'):
                std *= (2 * self.config.n_layer) ** -0.5
            torch.nn.init.normal_(module.weight, mean=0.0, std=std)
            if module.bias is not None:
                torch.nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, idx, targets=None, start_pos=0):
        B, T = idx.size()
        assert T <= self.config.block_size, f"Sequence length {T} exceeds block size {self.config.block_size}"

        # Token embeddings only (no positional embeddings)
        x = self.transformer.wte(idx)  # (B, T, n_embd)

        # Slice freqs_cis for current sequence positions
        freqs_cis = self.freqs_cis[start_pos:start_pos + T]

        # Forward through transformer blocks
        for block in self.transformer.h:
            x = block(x, start_pos, freqs_cis)  # mask=None, handled by MLA’s causal attention

        # Final normalization and projection
        x = self.transformer.ln_f(x)
        logits = self.lm_head(x)  # (B, T, vocab_size)

        loss = None
        if targets is not None:
            loss = F.cross_entropy(logits.view(-1, logits.size(-1)), targets.view(-1))

        return logits, loss

    @classmethod
    def from_pretrained(cls, model_type):
        assert model_type in {'gpt2', 'gpt2-medium', 'gpt2-large', 'gpt2-xl'}
        from transformers import GPT2LMHeadModel
        print(f"Loading weights from pretrained GPT: {model_type}")

        config_args = {
            'gpt2': dict(n_layer=12, n_head=12, n_embd=768),
            'gpt2-medium': dict(n_layer=24, n_head=16, n_embd=1024),
            'gpt2-large': dict(n_layer=36, n_head=20, n_embd=1280),
            'gpt2-xl': dict(n_layer=48, n_head=25, n_embd=1600),
        }[model_type]
        config_args['vocab_size'] = 50257
        config_args['block_size'] = 1024
        config_args['bias'] = False  # Set bias=False for consistency
        config = GPTConfig(**config_args)
        model = GPT(config)
        sd = model.state_dict()
        sd_keys = [k for k in sd.keys() if not k.endswith('.attn.bias')]

        model_hf = GPT2LMHeadModel.from_pretrained(model_type)
        sd_hf = model_hf.state_dict()
        sd_keys_hf = [k for k in sd_hf.keys() if not k.endswith('.attn.masked_bias') and not k.endswith('.attn.bias')]

        # Note: This assumes weight names align; DeepSeek-specific layers (MLA) may need custom mapping
        transposed = ['attn.c_attn.weight', 'attn.c_proj.weight', 'mlp.c_fc.weight', 'mlp.c_proj.weight']
        assert len(sd_keys_hf) == len(sd_keys), f"Mismatched keys: {len(sd_keys_hf)} != {len(sd_keys)}"
        for k in sd_keys_hf:
            if any(k.endswith(w) for w in transposed):
                assert sd_hf[k].shape[::-1] == sd[k].shape
                with torch.no_grad():
                    sd[k].copy_(sd_hf[k].t())
            else:
                assert sd_hf[k].shape == sd[k].shape
                with torch.no_grad():
                    sd[k].copy_(sd_hf[k])
        return model

# DataLoaderLite (unchanged)
class DataLoaderLite:
    def __init__(self, B, T):
        self.B = B
        self.T = T
        with open('/content/drive/MyDrive/datasets/input.txt', 'r') as f:  # Adjust path as needed
            text = f.read()
        enc = tiktoken.get_encoding('gpt2')
        tokens = enc.encode(text)
        self.tokens = torch.tensor(tokens)
        print(f"Loaded {len(self.tokens)} tokens")
        print(f"1 epoch = {len(self.tokens) // (B * T)} batches")
        self.current_position = 0

    def next_batch(self):
        B, T = self.B, self.T
        buf = self.tokens[self.current_position:self.current_position + B * T + 1]
        x = buf[:-1].view(B, T)
        y = buf[1:].view(B, T)
        self.current_position += B * T
        if self.current_position + (B * T + 1) > len(self.tokens):
            self.current_position = 0
        return x, y

# Training loop
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

torch.manual_seed(1337)
if torch.cuda.is_available():
    torch.cuda.manual_seed(1337)

train_loader = DataLoaderLite(B=8, T=512)

model = GPT(GPTConfig())
model.to(device)
model = torch.compile(model)  # Requires PyTorch 2.0+

max_lr = 6e-4
min_lr = max_lr * 0.1
warmup_steps = 10
max_steps = 50

def get_lr(it):
    if it < warmup_steps:
        return max_lr * (it + 1) / warmup_steps
    if it > max_steps:
        return min_lr
    decay_ratio = (it - warmup_steps) / (max_steps - warmup_steps)
    assert 0 <= decay_ratio <= 1
    coeff = 0.5 * (1.0 + math.cos(math.pi * decay_ratio))
    return min_lr + coeff * (max_lr - min_lr)

optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4, betas=(0.9, 0.95), eps=1e-8)
scaler = GradScaler(device=device)

for step in range(max_steps):
    t0 = time.time()
    x, y = train_loader.next_batch()
    x, y = x.to(device), y.to(device)
    optimizer.zero_grad()

    with autocast(device_type=device, dtype=torch.float16):
        logits, loss = model(x, targets=y)  # Explicitly use targets=

    scaler.scale(loss).backward()
    scaler.unscale_(optimizer)
    norm = torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
    lr = get_lr(step)
    for param_group in optimizer.param_groups:
        param_group['lr'] = lr
    scaler.step(optimizer)
    scaler.update()

    if torch.cuda.is_available():
        torch.cuda.synchronize()
    t1 = time.time()
    dt = (t1 - t0) * 1000
    tokens_per_sec = (train_loader.B * train_loader.T) / (t1 - t0)
    print(f"step {step:4d} | loss: {loss.item():.6f} | lr: {lr:.6f} | norm: {norm:.4f} | dt: {dt:.2f}ms | tok/sec: {tokens_per_sec:.2f}")
```

```python
def precompute_freqs_cis(seq_len, head_dim, device):
    theta = 10000.0  # Common base for rotary embeddings
    freqs = 1.0 / (theta ** (torch.arange(0, head_dim, 2, device=device, dtype=torch.float) / head_dim))
    t = torch.arange(seq_len, device=device, dtype=torch.float)
    freqs = torch.outer(t, freqs)  # Shape: (seq_len, head_dim / 2)
    freqs_cis = torch.polar(torch.ones_like(freqs), freqs)  # Complex tensor
    return freqs_cis
```
